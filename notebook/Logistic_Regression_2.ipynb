## Binary Classification probleam using Logistic Regression

### we use "student Placement dataset" for 'Logistic Regression'(Sigmoid function) 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df = pd.read_csv(r'D:\Yuvraj\My practice\ML\Dataset\Student_Placement_data.csv')
```

```python
df.head()
```

```python
# by using matplotlib we can plot the point on graph for better understanding

plt.scatter(df.cgpa,df.iq,df.placed,marker='+',color='green')

plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('CGPA vs IQ (Placed)')
plt.show()
```

```python
# define x,y for model building

X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
X_test.shape
```

```python
y_train
```

```python
y_test.shape
```

```python
# We can call LogisticRegression model

model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_pred = model.predict(X_test)
```

```python
# we can calculate the accuracy score by using the folloeing command

print(accuracy_score(y_test,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```
